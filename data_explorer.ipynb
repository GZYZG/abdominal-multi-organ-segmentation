```python
import SimpleITK as sitk
import skimage.io as io
import os

def read_img(path):
 img = sitk.ReadImage(path)
 data = sitk.GetArrayFromImage(img)
 return data
#显示一个系列图
def show_img_series(data):
 for i in range(data.shape[0]):
  io.imshow(data[i,:,:], cmap = 'gray')
  print(i)
  io.show()
 
#单张显示
def show_img(ori_img):
 io.imshow(ori_img[100], cmap = 'gray')
 io.show()
```

```python
#window下的文件夹路径 
path = './train/CT/'
label_path = './train/GT/'
samples = os.listdir(path)
samples = list(map(lambda x: path+x, samples))
labels = os.listdir(label_path)
labels = list(map(lambda x: label_path+x, labels))

sample = samples[0]
label = labels[0]

cts = read_img(sample)
segs = read_img(label)
```

```python
show_img_series(cts)
```

```python
show_img_series(segs)
```
